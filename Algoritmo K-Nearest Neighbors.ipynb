```python
import pandas as pd
# import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/TASS2019_country_CR_train.csv')
```

```python
df = df[df['sentiment'].notnull()]
df
```

```python
df['sentiment'] = df.sentiment.map({'N':0, 'P':1, 'NEU':2, 'NONE':3})
df
```

```python
len_str = [len(i) for i in df['content']]
df.insert(4,"Len",len_str,True)
len_str
```

```python
negative = df[df['sentiment'] == 0]
negative
```

```python
positive = df[df['sentiment'] == 1]
positive
```

```python
neutral = df[df['sentiment'] == 2]
neutral
```

```python
none = df[df['sentiment'] == 3]
none
```

```python
plt.scatter(negative['Unnamed: 0'], negative['Len'],
            marker="+", s=150, color="red",
            label="Negative (Class: 0)")
plt.scatter(positive['Unnamed: 0'], positive['Len'],
            marker="+", s=150, color="green",
            label="Positive (Class: 1)")
plt.scatter(neutral['Unnamed: 0'], neutral['Len'],
            marker="+", s=150, color="skyblue",
            label="Neutral (Class: 2)")
plt.scatter(none['Unnamed: 0'], none['Len'],
            marker="+", s=150, color="orange",
            label="None (Class: 3)")

plt.xlabel("User ID")
plt.ylabel("Length")
plt.legend(bbox_to_anchor=(1,0.2))
plt.show()
```

```python
data = df[['Unnamed: 0', 'Len']].astype(int)
sentiment = df['sentiment']

from sklearn import preprocessing
scale = preprocessing.MinMaxScaler()
scale.fit_transform(data)
data
```

```python
# Dividir los datos en conjunto de entrenamiento y de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, sentiment, random_state=4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
```
